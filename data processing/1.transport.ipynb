```python
from glob import glob
import shutil

li = glob('./Bbox_16_new/*')

for i in li:    # i : zip 
    if i[2:6] == 'Bbox':    # Bbox  
        lli = glob(i+'/*')  # lli : zip 
        for j in li:
            llli = glob(j+'/*')
            for k in llli:
                dir = './imgs'
                print(k)
                shutil.move(k,dir)
    
```
